## Fit various models and check their prediction of the specimen

```python
print("True class:", y_specimen)
```

```python
#X_train[:, 0].shape
#X_test[:, 0].shape
#x_specimen[0].shape

tsfs = [ex.classifiers.TimeSeriesForestClassifier(n_estimators=100, n_jobs=-1, random_state=1) for i in range(num_dims)]

print("True class:", y_specimen)

preds = []

for i in range(num_dims):
    tsfs[i].fit(X_train[:, i], y_train)
    preds.append(tsfs[i].predict(x_specimen[i][np.newaxis])[0])
    print("TSF prediction dim_" + str(i) + ": ", preds[i])


for i in range(num_dims):
    print("TSF accuracy score dim_" + str(i) + ": ", tsfs[i].score(X_test[:, i], y_test))
```

```python
slice = 10

time_om = tx.om.TimeSliceOmitter(size_x, time_slicing=slice, x_repl=tx.om.x_sample)

tsfs_expl_agnostic = [tx.om.KernelShapExplainer(time_om, tsfs[i].predict_proba, X_bg=X_test[:, i], y_bg=y_test,
                                                  n_samples=500, n_builds=5, bgcs=True).explain(x_specimen[i]) for i in range(num_dims)]
```

```python
def plot1d(ax, expl, title):
    ax.set_title(title)
    tx.plot.saliency1d(expl, ax=ax, model_output=y_specimen, slice_edge_style=None, legend_style=None, impact_legend=None)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    

#plt.title("BasicMotions")
fig, axes = plt.subplots(num_dims, 1, figsize=(20, 20))

bounds = [np.max(np.abs(tsfs_expl_agnostic[i].impacts[y_specimen])) for i in range(num_dims)]

impact_bound = max(bounds)

print("impact_bound: ", impact_bound)

for i in range(num_dims):
    #tx.plot.saliency1d(tsfs_expl_agnostic[i], model_output=y_specimen, ax=axes[i], impact_bound=impact_bound, 
                       #legend_style=None, impact_legend=None)
    plot1d(axes[i], tsfs_expl_agnostic[i], "TSF_dim_" + str(i))
    

#tx.plot.saliency2d(sktime_weasel_expl_agnostic, model_output=y_specimen, ax=ax2, impact_bound=impact_bound, legend_style=None)
#plot2d(ax2, sktime_weasel_expl_agnostic, "WEASEL by model-agnostic explainer")

for i in range(num_dims):
    axes[i].margins(y=0.15)
    #axes[i].set_title("TSF_dim_" + str(i))
    
fig.tight_layout()
#plt.savefig(ds + "_slice=" + str(slice) + "_true_class=" + str(y_specimen) + "_dims_preds=" + str(preds) +  ".png")
```
